```python
# @title 4. Gerar V√≠deos de Heatmap (Heatmap RGB e Heatmap MONOCROM√ÅTICO)
import glob
import numpy as np
import cv2
import matplotlib.pyplot as plt

def create_heatmap_video(npz_path, output_video_name, fps=30, colormap=cv2.COLORMAP_INFERNO):
    # Carregar dados
    data = np.load(npz_path)
    key = list(data.keys())[0]
    depth_data = data[key]

    # Se for arquivo √∫nico empacotado (Frames, H, W)
    if depth_data.ndim == 3:
        frames = depth_data
    else:
        frames = np.expand_dims(depth_data, axis=0)

    h, w = frames.shape[1], frames.shape[2]

    # Configurar VideoWriter
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video_name, fourcc, fps, (w, h))

    print(f"   -> Gerando {output_video_name} com {len(frames)} frames...")

    for i in range(len(frames)):
        d = frames[i]

        # Normalizar a profundidade para o range 0-255
        d_min = d.min()
        d_max = d.max()
        if d_max - d_min > 0:
            d_norm = (d - d_min) / (d_max - d_min)
        else:
            d_norm = d

        d_uint8 = (d_norm * 255).astype(np.uint8)

        # Aplicar Colormap escolhido (INFERNO para cor, BONE para cinza)
        heatmap = cv2.applyColorMap(d_uint8, colormap)

        out.write(heatmap)

    out.release()

# Encontrar os arquivos .npz
files_rgb = sorted(glob.glob("outputs_rgb/**/*.npz", recursive=True))
files_gray = sorted(glob.glob("outputs_gray/**/*.npz", recursive=True))

if files_rgb and files_gray:
    print("üé® Renderizando v√≠deos de Heatmap...")

    # Heatmap Colorido (Padr√£o) para o resultado RGB
    create_heatmap_video(files_rgb[0], "Heatmap_RGB_Color.mp4", fps=30, colormap=cv2.COLORMAP_INFERNO)

    # Heatmap Monocrom√°tico (Escala de Cinza) para o resultado Grayscale
    create_heatmap_video(files_gray[0], "Heatmap_Gray_Monochromatic.mp4", fps=30, colormap=cv2.COLORMAP_BONE)

    print("‚úÖ V√≠deos de Heatmap criados!")
else:
    print("‚ùå Arquivos NPZ n√£o encontrados.")
```

```python
# @title 5. An√°lise de Loss e Download
import matplotlib.pyplot as plt
from google.colab import files

print("üìä Calculando Loss e gerando relat√≥rio...")

# Carregar dados novamente para c√°lculo
d_r = np.load(files_rgb[0])[list(np.load(files_rgb[0]).keys())[0]]
d_g = np.load(files_gray[0])[list(np.load(files_gray[0]).keys())[0]]

# Garantir formato de v√≠deo (3 dimens√µes)
if d_r.ndim == 2: d_r = np.expand_dims(d_r, axis=0)
if d_g.ndim == 2: d_g = np.expand_dims(d_g, axis=0)

min_frames = min(len(d_r), len(d_g))
mse_losses = []

for i in range(min_frames):
    # Normalizar antes de comparar (Importante pois escalas podem variar)
    f_r = (d_r[i] - d_r[i].min()) / (d_r[i].max() - d_r[i].min())
    f_g = (d_g[i] - d_g[i].min()) / (d_g[i].max() - d_g[i].min())

    mse = np.mean((f_r - f_g) ** 2)
    mse_losses.append(mse)

# Plotar
plt.figure(figsize=(10, 5))
plt.plot(mse_losses, color='red', linewidth=2, label='Diferen√ßa (RGB vs Cinza)')
plt.title(f"Gr√°fico de Loss (Robustez √† Cor)")
plt.xlabel("Frame")
plt.ylabel("Erro M√©dio (MSE Normalizado)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig("Grafico_Loss.png")
plt.show()

print(f"‚úÖ Loss M√©dia Final: {np.mean(mse_losses):.5f}")

# Download
print("üì¶ Baixando arquivos...")
files.download('Heatmap_RGB_Color.mp4')
files.download('Heatmap_Gray_Monochromatic.mp4')
files.download('Grafico_Loss.png')
```
